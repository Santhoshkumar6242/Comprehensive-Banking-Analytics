### Data preprocessing.

Here we can drop the irrelevant features like ID, Customer_ID, Name, SSN.
and separate the features into numerical & categorical features for feature selection.

Standardize and Encode the numeric & categorical features by using the standartScaler & OneHotEncoder.



```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer


# Drop irrelevant columns
data_cleaned = data.drop(columns=['ID', 'Customer_ID', 'Name', 'SSN'])

# Select features for clustering
numerical_features = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 
                      'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Credit_History_Age', 
                      'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance']

categorical_features = ['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Credit_Score']

# One-hot encode categorical features and standardize numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Preprocess the data
features_preprocessed = preprocessor.fit_transform(data_cleaned)

# Convert to DataFrame for better visualization (optional)
features_preprocessed_df = pd.DataFrame(features_preprocessed)

# Check the first few rows of the preprocessed data
features_preprocessed_df.head()
```

### Determine the Optimal cluster number.

After perprocessing the data, determine the optimal cluster number for Kmeans algorithm.
Here I have used Elbow method and silhouette_score to find the accqurate metrics (optimal cluster number) for clustering.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Elbow Method
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(features_preprocessed)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()

# Silhouette Score
silhouette_scores = []
for n in range(2, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(features_preprocessed)
    score = silhouette_score(features_preprocessed, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()
```

From the plot, optimal cluster number was determined k= 4. approximately it takes 30 min fo run time so be patient.

### Apply K-Means clustering 

Once you have determined the optimal number of clusters, you can apply the K-Means algorithm to segment your customers.

```python
# Optimal number of clusters (update this based on Elbow and Silhouette methods)
optimal_clusters = 4

# K-Means clustering
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(features_preprocessed)

# Add cluster labels to original data
data_cleaned['cluster'] = cluster_labels
```

### Analyze and Interpret the Clusters
